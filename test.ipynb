```python
import json
from datasets import load_dataset

with open("/home/xulei/shayan/diversity/seed_cond/results/AIME2024/Qwen3-4B_0.6_1024_0.json", 'r') as f:
    results = json.load(f)["results"]
    
ds = load_dataset("Maxwell-Jia/AIME_2024")
df = ds["train"].to_pandas()
```

```python
def build_prompt(question, ref, response):
    return ("You are a rigorous grading instructor for math contest solutions."
            "You are given the original problem text, the student’s full written "
            "solution (whose final numeric answer is known to be correct), and one "
            "or more reference solutions (provided only as guidance; exact verbatim "
            "matching is NOT required). Your task is to judge whether the student’s "
            "*reasoning process* is logically valid and self-contained, instead of "
            "merely checking the final answer.\n\n[Original problem]"
            f"\n\n{question}\n\n"
            "[Student’s full solution]"
            f"\n\n{response}\n\n"
            "[Reference solution(s) - for guidance only]"
            f"\n\nSolution 1: {ref}\n\n"
            "Judging criteria:\n"
            "1) Are key derivations justified with sufficient intermediate steps, without "
            "circular reasoning or illicit assumptions?\n"
            "2) Minor arithmetic/notation slips that are explicitly corrected later and "
            "do not affect the logic may still be acceptable.\n"
            "3) If there is a fundamental flaw (misused theorem, false equality, missing essential conditions) "
            "such that the correct final answer could be a coincidence, the solution should be judged incorrect.\n"
            "4) Different approaches from the reference are permitted as long as the argument forms a logically "
            "sound and complete proof. Please reason step by step before you decide.\n"
            "Output requirement:\n"
            "After your analysis, the VERY LAST LINE must be exactly one of the following:\n"
            "[CORRECT]\n"
            "or\n"
            "[INCORRECT]\n"
            "Do not output anything after that final line.")
```

```python
s = 0
incorrect = [15, 22]
t = ""
for r in results:
    true_indices = [i for i, val in enumerate(r['correctness']) if val]
    if len(true_indices) > 0 and int(r['id']) not in incorrect:
        s +=1
        print(f"Question ID: {r['id']}")
        print(f"Num Correct Answers: {len(true_indices)}")
        t = t + str(r['id']) + ", "
    

print(f"Total Correct Answers: {s} ({s/30})")
print("Correctly Answered Solutions: ", t[:-2])
    
```
